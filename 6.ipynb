```python
import os
import sys
import numpy as np
import zipfile
import urllib.request
```

```python
# Find the hidden .zip file (:-)
urllib.request.urlretrieve(
    'http://www.pythonchallenge.com/pc/def/channel.zip',
    'data/channel.zip');
```

```python
data = zipfile.ZipFile('data/channel.zip')
data.filelist
```

```python
# Extract all the files into a given folder
data.extractall('data/channel/')
```

```python
# Read the readme.txt file
with open('data/channel/readme.txt') as f:
    print(f.read())
```

```python
# Loop over the files following the hint
n = 90052
numbers = [str(n)]
total = 910 # number of files in the zip
for i in range(1, total):
    with open('data/channel/{}.txt'.format(n)) as f:
        text = f.read()
        print('{}: {}'.format(i, text))
        n = text[16:len(text)]
        try:
            n_ = int(n)
            numbers.append(n)
        except ValueError:
            print("Not a number")
            break
```

```python
# Loop over the files catching the comments
n = 90052
comments = ""
numbers = [str(n)]
total = 910 # number of files in the zip
for i in range(1, total):
    with open('data/channel/{}.txt'.format(n)) as f:
        text = f.read()
        comment = data.getinfo('{}.txt'.format(n)).comment
        comments += comment.decode()
        n = text[16:len(text)]
        try:
            n_ = int(n)
            numbers.append(n)
        except ValueError:
            print("Not a number")
            break
print(comments)
```
