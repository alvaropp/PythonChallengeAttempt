```python
import urllib.request
import urllib.parse
import pickle
```

```python
# Obtain the address of the banner.p file which appears in the source code
address = urllib.parse.urljoin('http://www.pythonchallenge.com/pc/def/peak.html', 'banner.p')
```

```python
# Download it to banner.p file
with urllib.request.urlopen(address) as response, open('data/banner.p', 'wb') as out_file:
    data = response.read()
    out_file.write(data)
```

```python
with open('data/banner.p', 'rb') as banner:
    content = pickle.load(banner)
```

```python
for line in content:
    print(line)
    print('-'*70)
```

```python
# All lines sum up to the same number 95, there are two characters: ' ' & '#'
for line in content:
    print('line sum = {}'.format(sum([line[i][1] for i in range(len(line))])))
```

```python
# Print the specified characters the number of times given by the associated numbers
for line in content:
    print("".join([line[i][0]*line[i][1] for i in range(len(line))]))
```
