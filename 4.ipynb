```python
import urllib.request
```

```python
nothing = 12345
nothingList = [nothing]
for i in range(400):
    web = urllib.request.urlopen('http://www.pythonchallenge.com/pc/def/linkedlist.php?nothing={}'.format(nothing))
    source = web.read()
    print(i, source)
    try:
        nothing = int(source.split()[-1])
    except:
        nothing = input("MANUALLY INTRODUCE NEXT NOTHING OR 'CANCEL': ")
    if nothing == 'CANCEL':
        break
    nothingList.append(nothing)
    # Check for repeated values, just in case we get trapped in a closed loop
    if len(set(nothingList)) != len(nothingList):
        print("========== REPEATED NOTHING VALUE APPEARED ==========")
        break
```
