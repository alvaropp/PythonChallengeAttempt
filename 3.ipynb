```python
import urllib.request
```

```python
web = urllib.request.urlopen('http://www.pythonchallenge.com/pc/def/equality.html')
source = web.read()
text = source.decode().split('<!--')[1]
text = "".join(text.split('\n'))
```

```python
print(text)
```

```python
letters = ""
for i in range(3, len(text) - 3):
    if text[i].islower():
        if text[i - 1].isupper():
            if text[i - 2].isupper():
                if text[i - 3].isupper():
                    if text[i - 4].islower():
                        if text[i + 1].isupper():
                            if text[i + 2].isupper():
                                if text[i + 3].isupper():
                                    if text[i + 4].islower():
                                        letters += text[i]

print(letters)
```
